# Importing the Libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

# EDA

## Sepal Length Count

```python
ax = sns.countplot(x="sepal length (cm)", data=dataset)
sns.set(rc={"figure.figsize": (25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

## Sepal Width Count

```python
ax = sns.countplot(x="sepal width (cm)", data=dataset)
sns.set(rc={"figure.figsize": (25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

## Petal Length Count

```python
ax = sns.countplot(x="petal length (cm)", data=dataset)
sns.set(rc={"figure.figsize": (25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

## Petal Width Count

```python
ax = sns.countplot(x="petal width (cm)", data=dataset)
sns.set(rc={"figure.figsize": (25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

## Species Count

```python
ax = sns.countplot(x="Species", data=dataset)
sns.set(rc={"figure.figsize": (8,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

## Scatter Plot of Sepal Length vs Sepal Width

```python

sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='Species', data=dataset)
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
```

## Scatter Plot of Petal Length vs Petal Width

```python
sns.scatterplot(x="petal length (cm)", y="petal width (cm)", hue="Species", data=dataset)
plt.title('Scatter Plot of Petal Length vs Petal Width')
plt.show()
```

## Scatter Plot of Sepal Length vs Petal Length

```python
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="Species", data=dataset)
plt.title("Scatter Plot of Sepal Length vs Petal Length")
plt.show()
```

## Scatter Plot of Sepal Width vs Petal Width

```python
sns.scatterplot(x="sepal width (cm)", y="petal width (cm)", hue="Species", data=dataset)
plt.title("Scatter Plot of Sepal Width vs Petal Width")
plt.show()
```

## Box Plot of Sepal Length/Sepal Width by Species

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
sns.boxplot(x='Species', y='sepal length (cm)', data=dataset)
plt.title('Box Plot of Sepal Length by Species')

plt.subplot(2, 1, 2)
sns.boxplot(x='Species', y='sepal width (cm)', data=dataset)
plt.title('Box Plot of Sepal Width by Species')

plt.tight_layout()
plt.show()
```

## Box Plot of Petal Length/Box Width by Species

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
sns.boxplot(x='Species', y='petal length (cm)', data=dataset)
plt.title('Box Plot of Petal Length by Species')

plt.subplot(2, 1, 2)
sns.boxplot(x='Species', y='petal width (cm)', data=dataset)
plt.title('Box Plot of Petal Width by Species')

plt.tight_layout()
plt.show()
```

## Violin Plot of Sepal Length/Sepal Width by Species

```python
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
sns.violinplot(x='Species', y='sepal length (cm)', data=dataset)
plt.title('Violin Plot of Sepal Length by Species')

plt.subplot(2, 1, 2)
sns.violinplot(x='Species', y='sepal width (cm)', data=dataset)
plt.title('Violin Plot of Sepal Width by Species')

plt.tight_layout()
plt.show()
```

## Violin Plot of Petal Length/Petal Width by Species

```python
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
sns.violinplot(x='Species', y='petal length (cm)', data=dataset)
plt.title('Violin Plot of Petal Length by Species')

plt.subplot(2, 1, 2)
sns.violinplot(x='Species', y='petal width (cm)', data=dataset)
plt.title('Violin Plot of Petal Width by Species')

plt.tight_layout()
plt.show()
```

## Pair Plot

```python
sns.pairplot(dataset, hue='Species', height=2.5)
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```
